GroupBy

```python
df['drive-wheels'].unique()
```

```python
group_DBP=df[['drive-wheels','body-style','price']]
```

```python
gp=group_DBP.groupby(['drive-wheels'], as_index=False).mean()
```

```python
gp
```

1. groupby drive-wheels and body-style

```python
gp1=group_DBP.groupby(['drive-wheels','body-style'], as_index=False).mean()
```

```python
gp1
```

```python
gp1_pivot=gp1.pivot(index='drive-wheels', columns='body-style')
gp1_pivot
```

```python
gp1_pivot.fillna(0)
```

```python
#heat map to visualise relationship b/w body-style and price
plt.pcolor(gp1_pivot, cmap="RdBu")
plt.colorbar()
plt.show()
```

```python
fig, ax = plt.subplots()
im = ax.pcolor(gp1_pivot, cmap='RdBu')

#label names
row_labels = gp1_pivot.columns.levels[1]
col_labels = gp1_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(gp1_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(gp1_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```
